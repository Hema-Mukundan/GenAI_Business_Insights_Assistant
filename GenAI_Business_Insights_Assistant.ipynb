**Load Dataset to Python**

```python
import pandas as pd

df = pd.read_excel("/Superstore.xlsx")
df.head()
```

**Data Pre-processing**

```python
df.info()
```

```python
df.isnull().sum()
```

**Drop column 'Postal Code'**

```python
df = df.drop(columns=['Postal Code'])
```

**Business Summaries**

```python
# Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum()

# Profit by Category
profit_by_category = df.groupby('Category')['Profit'].sum()

# Top 5 Sub-Categories by Sales
top_subcategory = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(5)

# Profit by Segment
profit_by_segment = df.groupby('Segment')['Profit'].sum()
```

**Convert Data → AI Input**

```python
summary_text = f"""
Sales by Region:
{sales_by_region}

Profit by Category:
{profit_by_category}

Top Sub-Categories:
{top_subcategory}

Profit by Segment:
{profit_by_segment}
"""
```

```python
print(summary_text)
```
